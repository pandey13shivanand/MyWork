**Q1: Write a Python Program to Check if a Number is Positive, Negative or Zero?**

```python
user_input = int(input("Enter the int value :"))

if user_input > 0:
    print("User entered positive number")
elif user_input < 0:
    print("User entered negative number")
else:
    print("User entered zero")
```

**Q2.Write a Python Program to Check if a Number is Odd or Even?**

```python
user_input = int(input("Enter the int value :"))

if user_input == 0:
     print("User entered zero")
elif user_input//2 == 1 or user_input%2 == 0:
     print("User enetered Even number")
else:
    print("User entered Odd number")
```

**Q3.Write a Python Program to Check Leap Year?**

```python
year_input = int(input("Enter the year to check :"))

if year_input%400 == 0 or year_input%100==0 and year_input%4==0:
    print("The given year is Leap Year")
else:
    print("The given year is not a Leap year")
```

**Q4.Write a Python Program to Check Prime Number?**

```python
user_input = int(input("Enter a number : "))

if user_input >1:
    for i in range(2,int(user_input/2)+1):
        if user_input%i == 0:
            print("The number is not Prime")
            break;
    else:
         print("The number is Prime")
else:
    print("The number is not Prime")
```
